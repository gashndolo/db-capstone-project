```python
### Lsts find orders greater than Ksh1900 coz my dummy data is in Kenyan Shillings
orders_greater = """
    SELECT 
        CONCAT(c.FirstName, ' ', c.LastName) AS FullName,
        c.Email,
        c.PhoneNumber
    FROM customers c
    JOIN orders o ON c.CustomerID = o.CustomerID
    WHERE o.TotalCost > 1900;
    """
try: 
    cursor.execute(orders_greater)
    print("Fetching results")
    results = cursor.fetchall()
    for result in results:
        print(result)
except:
    print("ERROR: Something went wrong")
```
